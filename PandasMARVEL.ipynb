```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset with 4 subjects as per your input
data = {
    'Subject': ['ME', 'EE', 'M1', 'PH'],
    'Hours_Studied': [35, 42, 45, 30],
    'Marks_Scored': [87, 88, 95, 91],
    'Attendance (%)': [88, 85, 98, 92]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Plot 1: Line graph (Hours Studied and Marks Scored per Subject)
plt.figure(figsize=(10, 6))
plt.plot(df['Subject'], df['Hours_Studied'], label='Hours Studied', marker='o', linestyle='-', color='blue')
plt.plot(df['Subject'], df['Marks_Scored'], label='Marks Scored', marker='s', linestyle='--', color='orange')
plt.title('Study Hours and Marks Scored by Subject', fontsize=14)
plt.xlabel('Subject', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()

# Plot 2: Bar graph (Attendance Percentage per Subject)
plt.figure(figsize=(10, 6))
plt.bar(df['Subject'], df['Attendance (%)'], color='green', alpha=0.7)
plt.title('Attendance Percentage by Subject', fontsize=14)
plt.xlabel('Subject', fontsize=12)
plt.ylabel('Attendance (%)', fontsize=12)
plt.show()

# Plot 3: Scatter plot (Hours Studied vs Marks Scored)
plt.figure(figsize=(10, 6))
plt.scatter(df['Hours_Studied'], df['Marks_Scored'], c='purple', alpha=0.8, edgecolors='k', s=100)
plt.title('Hours Studied vs Marks Scored', fontsize=14)
plt.xlabel('Hours Studied', fontsize=12)
plt.ylabel('Marks Scored', fontsize=12)
plt.grid(True)
plt.show()
```
